ESte script no tiene nada especial, es meramente un truncado de los datos para predecir, ya que solo necesitamos las últimas 72 horas

```python
import pandas as pd
```

```python
df = pd.read_parquet("../../data/final_data.parquet")
df
```

```python
# 1. Asegurar que el timestamp sea tipo datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 2. Ordenar por las llaves y el tiempo para que tail() tome los registros más recientes
df = df.sort_values(by=['VendorID', 'PULocationID', 'timestamp'])

# 3. Agrupar y tomar las últimas 72 filas (horas) por cada combinación
df_filtrado = df.groupby(['VendorID', 'PULocationID']).tail(72)
```

```python
# Guardar como Parquet con compresión Snappy
df_filtrado.to_parquet('data/recent_data.parquet', engine='pyarrow', compression='snappy')
```

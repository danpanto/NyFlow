```python
import polars as pl
import numpy as np
from pathlib import Path
import os
from datetime import datetime
from dateutil.relativedelta import relativedelta
from time import sleep
```

```python
path = Path(os.getcwd())
if not Path(path, "data").exists(): path = path.parent
```

```python
url_zone_lookup = "https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv"
df_zone_lookup = pl.read_csv(url_zone_lookup)
df_zone_lookup.head()
```

```python
def download(url_raw: str, year: int, month: int, name: str, overwrite = False) -> pl.LazyFrame | None:
    folder =  path / "data" / str(year) / str(month)
    out_path = folder / f"{name}.parquet"
    folder.mkdir(parents=True, exist_ok=True)

    if out_path.exists() and not overwrite: return None

    url = url_raw.format(year=year, month=month)

    pl.scan_parquet(url).sink_parquet(out_path)
```

```python
n_years = 5
years = np.repeat(np.arange(2026 - n_years, 2026),12)
months = np.tile(np.arange(1, 13), n_years)
```

```python
url_parquet_yellow = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_{year}-{month:02d}.parquet"

falla_fetch_yellow = []

for y,m in zip(years, months):
    y = int(y)
    m = int(m)
    try: 
        download(url_parquet_yellow, y, m, "yellow_taxi")
    except Exception as e:
        falla_fetch_yellow.append((y,m,str(e)))
```
